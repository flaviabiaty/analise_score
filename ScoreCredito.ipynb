Análise de Score de clientes de um banco, para definir se haverá disponibilização de empréstimos, crédito e outros benefícios.

##Importação e tratamento de dados

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/flaviabiaty/analise_score/refs/heads/main/clientes.csv')
df.head()
```

```python
df.info()
df.columns
```

algumas colunas estão com o tipo de informação “object”. Isso na verdade é um texto,
só que os modelos de classificação não conseguem trabalhar com
textos.
Então nós vamos precisar fazer um tratamento nessas colunas antes de
continuar com o nosso projeto.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#transformar colunas de texto em numeros
codificador = LabelEncoder()

#transformar todas as colunas exceto score_credito
for coluna in df.columns:
    if df[coluna].dtype == 'object' and coluna != 'score_credito':
        df[coluna] = codificador.fit_transform(df[coluna])

#verifica se foram modificadas
for coluna in df.columns:
    if df[coluna].dtype == 'object' and coluna != 'score_credito':
        print(coluna)
```

```python
df.head()
```

##Informações para treinamento do modelo

Treinamento do modelo de classificação: a partir das informações da base, separamos em treino e teste, para que o modelo
consiga fazer as previsões.

```python
#escolha das colunas para treinar o modelo
# y é a coluna que quero que o modelo calcule
# x são todas as colunas que serão utilizadas para prever o score de crédito (exceto id_cliente)
x=df.drop(['score_credito','id_cliente'],axis=1)
y=df['score_credito']

from sklearn.model_selection import train_test_split
x_treino,x_teste,y_treino,y_teste = train_test_split(x,y,test_size=0.3, random_state=1)
```

A Acurácia é uma das métricas de avaliação que utilizamos para verificar se um modelo é bom em relação a outro. A acurácia nada mais é do que a performance do modelo, ou seja, vai verificar quantas informações foram classificadas de forma
correta.

Quando trabalhamos em problemas de classificação vamos utilizar pelo menos 2 modelos para verificar qual deles tem o melhor resultado.

```python
#se o modelo considerasse "Standard para todos, qual seria a acurácia?"
contagem_score = df['score_credito'].value_counts()
print('Se considerasse todos Standard, a acurácia seria de: ',contagem_score['Standard']/sum(contagem_score))
print('Se considerasse todos Poor, a acurácia seria de: ',contagem_score['Poor']/sum(contagem_score))
print('Se considerasse todos Good, a acurácia seria de: ',contagem_score['Good']/sum(contagem_score))
```

```python
# Árvore de Decisão -> RandomForest
from sklearn.ensemble import RandomForestClassifier
# Vizinhos Próximos -> KNN
from sklearn.neighbors import KNeighborsClassifier

# criar modelo
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()
# treinar modelo
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

Utiliza-se o .predict com cada um dos modelos para fazer a previsão.
Feita a previsão, calcula-se a métrica de acurácia para verificar quanto cada um
dos modelos acertou.

```python
# previsoes
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)
#acuracia
from sklearn.metrics import accuracy_score
print('A acurácia para o Modelo de árvore de decisão:')
display(accuracy_score(y_teste, previsao_arvoredecisao))
print('A acurácia para o Modelo KNN:')
display(accuracy_score(y_teste, previsao_knn))
```

Neste caso, o modelo de árvore de decisão apresentou um resultado melhor de precisão (82%). Então será utilizado para previsão de score dos clientes

Além de definir qual o modelo é melhor para fazer a classificação, o modelo pode ser usado para verificar quais as características mais importantes para definir o score de crédito.

```python
colunas = list(x_teste.columns)
importancia = pd.DataFrame(index=colunas,data=modelo_arvoredecisao.feature_importances_,columns=['Importância(%)'])
importancia = importancia * 100
importancia.sort_values(by='Importância(%)',ascending=False)
```
